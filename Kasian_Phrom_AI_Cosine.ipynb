# Hybrid TF-IDF รวมกับ Feature Vector

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity



def create_tfidf_embeddings(nursing_houses):
    descriptions = [
        f'{nh["Address"]}' for index, nh in nursing_houses.iterrows()
    ]
    # vectorizer = TfidfVectorizer()
    vectorizer = TfidfVectorizer(lowercase=True, strip_accents='unicode')

    tfidf_embeddings = vectorizer.fit_transform(descriptions).toarray()
    return tfidf_embeddings


def create_feature_embeddings(nursing_houses):
    prices = np.array([nh["price"] for index, nh in nursing_houses.iterrows()]).reshape(-1, 1)
    latitudes = np.array([nh["latitude"] for index, nh in nursing_houses.iterrows()]).reshape(-1, 1)
    longitudes = np.array([nh["longitude"] for index, nh in nursing_houses.iterrows()]).reshape(-1, 1)

    scaler = MinMaxScaler()
    feature_embeddings = scaler.fit_transform(np.hstack([prices, latitudes, longitudes]))
    return feature_embeddings


def create_hybrid_embeddings(nursing_houses):
    tfidf_embeddings = create_tfidf_embeddings(nursing_houses)
    feature_embeddings = create_feature_embeddings(nursing_houses)
    hybrid_embeddings = np.hstack([tfidf_embeddings, feature_embeddings])
    return hybrid_embeddings


house_embeddings = create_hybrid_embeddings(nursing_houses)


print("House Embeddings Shape:", house_embeddings.shape)
print("Sample House Embedding:", house_embeddings[0])
```

```python

def calculate_cosine_similarity(house_embeddings):
    return cosine_similarity(house_embeddings)


def get_similar_nursing_homes_by_Name(nursing_houses, house_name, top_n=3):
    house_embeddings = create_hybrid_embeddings(nursing_houses)
    similarity_matrix = calculate_cosine_similarity(house_embeddings)

    
    query_index = nursing_houses[nursing_houses["Name"] == house_name].index

    if query_index.empty:
        return f"ไม่พบบ้านพักชื่อ '{house_name}'"

    query_index = query_index[0]


    similarity_scores = similarity_matrix[query_index]


    similar_houses = sorted(
        [(i, similarity_scores[i]) for i in range(len(nursing_houses)) if i != query_index],
        key=lambda x: x[1], reverse=True
    )[:top_n]

    recommendations = [
        {
            "Name": nursing_houses.iloc[i]["Name"].replace("\u200b", ""),
            "Address": nursing_houses.iloc[i]["Address"].replace("\u200b", ""),
            "price": nursing_houses.iloc[i]["price"],
            "similarity": round(score, 4)
        }
        for i, score in similar_houses
    ]

    return recommendations
```

```python
house_name = "บ้านเย็นจิต" #จังหวัด กาญจนบุรี มีบ้านหลังเดียวเพราะงั้นมันเลย มั่ว
recommendations = get_similar_nursing_homes_by_Name(nursing_houses, house_name, top_n=3)

print(f"บ้านพักที่คล้ายกับ '{house_name}'\n")
for rec in recommendations:
    print(rec)
```

# loop ชื่อบ้านเช็ค similarity score

```python
count = 0
for i in nursing_houses["Name"]:

  count += 1
  recommendations = get_similar_nursing_homes_by_Name(nursing_houses, i, top_n=3)

  province = nursing_houses.loc[nursing_houses["Name"] == i, "Province"].values[0]
  print(f"{count} บ้านพักที่คล้ายกับ '{i}' -> จังหวัด {province}")
  for rec in recommendations:
      print(rec)
  print("\n")

print(count)
#ขอนแก่น ส่วนใหญ่ได้ score ดี
```
